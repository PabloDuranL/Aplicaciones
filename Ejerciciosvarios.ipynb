#Sorts algorithms

```python
def burbuja(x):
  for j in range(len(x)-1):
    for i in range(len(x)-1):
      if x[i]>x[i+1]:
        x[i],x[i+1]=x[i+1],x[i]
  return x
```

```python
def insort(x):
  for i in range(len(x)-1):
    max=x[0]
    ind=0
    for j in range(len(x)-i):
       if x[j]>max:
        max=x[j]
        ind=j
    x[ind],x[-i-1]=x[-i-1],max
  return x
```

```python
def quicksort(x):
  if len(x)<=1:
    return x
  menor=[]
  igual=[]
  mayor=[]
  pivote=x[-1]
  for i in range(len(x)):
    if pivote > x[i]:
      menor.append(x[i])
    elif pivote < x[i]:
      mayor.append(x[i])
    else:
      igual.append(x[i])
  return quicksort(menor)+igual+quicksort(mayor)


```

```python
def factorial(n, dtype=int):
  if not isinstance(n,int) or n<0:
    return print("solo se puede numeros enteros")
  if n<=1:
    return 1
  prod=n
  return n*factorial(n-1)

print(factorial(6))
```

```python
def fibonnaci(n):
  if not isinstance(n,int) or n<0:
    return print("solo enteros")
  if n==0:
    return 0
  if n==1:
    return 1
  return fibonnaci(n-1)+fibonnaci(n-2)
```

```python
for i in range(15):
  print(f"El termino {i} de la serie es {fibonnaci(i)}")
```
